# Conduct a sensitivity analysis on the above results.

```r
library(dplyr)
library(ggplot2)
library(mclust)
library(pheatmap)
```

```r
# Place metadata with different bin sizes into a list.
meta_list <- list(
  Bin50  = Bin50.meta,
  Bin100 = Bin100.meta,
  Bin150 = Bin150.meta,
  Bin200 = Bin200.meta,
  Bin300 = Bin300.meta
)
names(meta_list)
```

```r
#### Define a function 2.1 Purity
calc_purity <- function(pred, truth) {
  pred <- as.factor(pred)
  truth <- as.factor(truth)
  keep <- !is.na(pred) & !is.na(truth)
  tab <- table(pred[keep], truth[keep])
  sum(apply(tab, 1, max)) / sum(tab)
}
####2.2 NMI
calc_nmi_manual <- function(x, y) {
  x <- as.factor(x)
  y <- as.factor(y)
  keep <- !is.na(x) & !is.na(y)
  x <- x[keep]
  y <- y[keep]
  
  tab <- table(x, y)
  n <- sum(tab)
  pxy <- tab / n
  px <- rowSums(pxy)
  py <- colSums(pxy)
  
  mi <- 0
  for (i in seq_len(nrow(pxy))) {
    for (j in seq_len(ncol(pxy))) {
      if (pxy[i, j] > 0) {
        mi <- mi + pxy[i, j] * log(pxy[i, j] / (px[i] * py[j]))
      }
    }
  }
  
  hx <- -sum(px[px > 0] * log(px[px > 0]))
  hy <- -sum(py[py > 0] * log(py[py > 0]))
  
  2 * mi / (hx + hy)
}
### 2.3 Spatial Continuity Metric: Here, we define a simple score—specifically, the proportion of the four immediate neighbors (up, down, left, and right) that belong to the same cluster.
calc_spatial_coherence <- function(df, cluster_col = "Bayes16") {
  df <- df %>%
    mutate(
      row = as.numeric(coor_x),
      col = as.numeric(coor_y),
      cluster = as.character(.data[[cluster_col]])
    ) %>%
    filter(!is.na(coor_x), !is.na(col), !is.na(cluster))
  
  # To prevent different slices from getting mixed up, calculate them separately based on `T_names`.
  score_list <- lapply(split(df, df$T_names), function(subdf) {
    key <- paste(subdf$row, subdf$col, sep = "_")
    cl_map <- setNames(subdf$cluster, key)
    
      neighbors <- rbind(
      cbind(subdf$coor_x + 1, subdf$col),
      cbind(subdf$coor_x - 1, subdf$col),
      cbind(subdf$coor_x, subdf$col + 1),
      cbind(subdf$coor_x, subdf$col - 1)
    )
    
    current_cluster <- rep(subdf$cluster, 4)
    neighbor_key <- paste(neighbors[,1], neighbors[,2], sep = "_")
    neighbor_cluster <- cl_map[neighbor_key]
    
    valid <- !is.na(neighbor_cluster)
    if (sum(valid) == 0) return(NA_real_)
    
    mean(current_cluster[valid] == neighbor_cluster[valid])
  }) 
  mean(unlist(score_list), na.rm = TRUE)
}
```

```r
#### 3）Consistency between Different Bin Sizes and the Manual Layer
binsize_eval <- data.frame()

for (nm in names(meta_list)) {
  df <- meta_list[[nm]]
  pred <- as.factor(df$Bayes16)
  truth <- as.factor(df$area)
  keep <- !is.na(pred) & !is.na(truth)
  
  binsize_eval <- rbind(binsize_eval, data.frame(
    bin_size = as.numeric(gsub("Bin", "", nm)),
    ARI = adjustedRandIndex(pred[keep], truth[keep]),
    NMI = calc_nmi_manual(pred, truth),
    Purity = calc_purity(pred, truth),
    SpatialCoherence = calc_spatial_coherence(df, "Bayes16")
  ))
}

binsize_eval <- binsize_eval[order(binsize_eval$bin_size), ]
binsize_eval
```

```r
## 4) Under-segmentation / Over-segmentation Analysis
### 4.1 How many clusters is each artificial layer divided into? The higher this number, the greater the tendency toward over-segmentation.
fragment_summary <- data.frame()

for (nm in names(meta_list)) {
  df <- meta_list[[nm]]
  
  tmp <- df %>%
    filter(!is.na(area), !is.na(Bayes16)) %>%
    group_by(area) %>%
    summarise(
      n_clusters_in_area = n_distinct(Bayes16),
      .groups = "drop"
    )
  
  fragment_summary <- rbind(fragment_summary, data.frame(
    bin_size = as.numeric(gsub("Bin", "", nm)),
    mean_clusters_per_area = mean(tmp$n_clusters_in_area),
    median_clusters_per_area = median(tmp$n_clusters_in_area)
  ))
}

fragment_summary <- fragment_summary[order(fragment_summary$bin_size), ]
fragment_summary
```

```r
## 4.2 The number of artificial layers injected into each cluster—the higher this number, the greater the tendency toward under-segmentation.
mix_summary <- data.frame()

for (nm in names(meta_list)) {
  df <- meta_list[[nm]]
  
  tmp <- df %>%
    filter(!is.na(area), !is.na(Bayes16)) %>%
    group_by(Bayes16) %>%
    summarise(
      n_area_in_cluster = n_distinct(area),
      .groups = "drop"
    )
  
  mix_summary <- rbind(mix_summary, data.frame(
    bin_size = as.numeric(gsub("Bin", "", nm)),
    mean_area_per_cluster = mean(tmp$n_area_in_cluster),
    median_area_per_cluster = median(tmp$n_area_in_cluster)
  ))
}

mix_summary <- mix_summary[order(mix_summary$bin_size), ]
mix_summary
```

```r
### 5）merge all
summary_df <- binsize_eval %>%
  left_join(fragment_summary, by = "bin_size") %>%
  left_join(mix_summary, by = "bin_size")

summary_df
```

```r
### 6）plot
#### 6.1 ARI / NMI / Purity / Spatial Continuity
p1 <- ggplot(summary_df, aes(x = bin_size, y = ARI)) +
  geom_line() +
  geom_point(size = 2) +
  theme_classic() +
  labs(x = "Bin size", y = "ARI with manual layers")

p2 <- ggplot(summary_df, aes(x = bin_size, y = NMI)) +
  geom_line() +
  geom_point(size = 2) +
  theme_classic() +
  labs(x = "Bin size", y = "NMI with manual layers")

p3 <- ggplot(summary_df, aes(x = bin_size, y = Purity)) +
  geom_line() +
  geom_point(size = 2) +
  theme_classic() +
  labs(x = "Bin size", y = "Purity with manual layers")

p4 <- ggplot(summary_df, aes(x = bin_size, y = SpatialCoherence)) +
  geom_line() +
  geom_point(size = 2) +
  theme_classic() +
  labs(x = "Bin size", y = "Spatial coherence")

print(p1)
print(p2)
print(p3)
print(p4)
```

```r
#### 6.2 Over-segmentation / Under-segmentation Trends
p5 <- ggplot(summary_df, aes(x = bin_size, y = mean_clusters_per_area)) +
  geom_line() +
  geom_point(size = 2) +
  theme_classic() +
  labs(x = "Bin size", y = "Mean clusters per manual layer")

p6 <- ggplot(summary_df, aes(x = bin_size, y = mean_area_per_cluster)) +
  geom_line() +
  geom_point(size = 2) +
  theme_classic() +
  labs(x = "Bin size", y = "Mean manual layers per cluster")

print(p5)
print(p6)
```

```r
#### 7) Calculate metrics separately for each section, then take the average (more robust).

### If you prefer not to pool all sections together directly—but instead wish to account for donor/section-specific variations—it is recommended that you perform calculations separately for each `T_name` and then aggregate the results.

### 7.1 ARI, NMI, and Purity scores for each `binsize` and each individual section.
```

```r
per_slice_eval <- data.frame()

for (nm in names(meta_list)) {
  df <- meta_list[[nm]]
  
  for (slice in unique(df$T_names)) {
    subdf <- df[df$T_names == slice, ]
    
    pred <- as.factor(subdf$Bayes16)
    truth <- as.factor(subdf$area)
    keep <- !is.na(pred) & !is.na(truth)
    
    if (sum(keep) > 10) {
      per_slice_eval <- rbind(per_slice_eval, data.frame(
        bin_size = as.numeric(gsub("Bin", "", nm)),
        T_names = slice,
        ARI = adjustedRandIndex(pred[keep], truth[keep]),
        NMI = calc_nmi_manual(pred, truth),
        Purity = calc_purity(pred, truth),
        SpatialCoherence = calc_spatial_coherence(subdf, "Bayes16")
      ))
    }
  }
}

per_slice_eval
```

```r
# 7.2 Summary: Mean ± SD
per_slice_summary <- per_slice_eval %>%
  group_by(bin_size) %>%
  summarise(
    ARI_mean = mean(ARI, na.rm = TRUE),
    ARI_sd = sd(ARI, na.rm = TRUE),
    NMI_mean = mean(NMI, na.rm = TRUE),
    NMI_sd = sd(NMI, na.rm = TRUE),
    Purity_mean = mean(Purity, na.rm = TRUE),
    Purity_sd = sd(Purity, na.rm = TRUE),
    SpatialCoherence_mean = mean(SpatialCoherence, na.rm = TRUE),
    SpatialCoherence_sd = sd(SpatialCoherence, na.rm = TRUE),
    n = n(),
    .groups = "drop"
  )
```

```r
#### ARI  mean ± SD
ggplot(per_slice_summary, aes(x = bin_size, y = ARI_mean)) +
  geom_line() +
  geom_point(size = 2) +
  geom_errorbar(aes(ymin = ARI_mean - ARI_sd, ymax = ARI_mean + ARI_sd), width = 10) +
  theme_classic() +
  labs(x = "Bin size", y = "ARI with manual layers")
```

```r
###NMI mean ± SD
ggplot(per_slice_summary, aes(x = bin_size, y = NMI_mean)) +
  geom_line() +
  geom_point(size = 2) +
  geom_errorbar(aes(ymin = NMI_mean - NMI_sd, ymax = NMI_mean + NMI_sd), width = 10) +
  theme_classic() +
  labs(x = "Bin size", y = "NMI with manual layers")
```

```r
### Purity mean ± SD
ggplot(per_slice_summary, aes(x = bin_size, y = Purity_mean)) +
  geom_line() +
  geom_point(size = 2) +
  geom_errorbar(aes(ymin = Purity_mean - Purity_sd, ymax = Purity_mean + Purity_sd), width = 10) +
  theme_classic() +
  labs(x = "Bin size", y = "Purity with manual layers")
```

```r
#### Spatial coherence mean ± SD
ggplot(per_slice_summary, aes(x = bin_size, y = SpatialCoherence_mean)) +
  geom_line() +
  geom_point(size = 2) +
  geom_errorbar(aes(ymin = SpatialCoherence_mean - SpatialCoherence_sd,
                    ymax = SpatialCoherence_mean + SpatialCoherence_sd), width = 10) +
  theme_classic() +
  labs(x = "Bin size", y = "Spatial coherence")
```
